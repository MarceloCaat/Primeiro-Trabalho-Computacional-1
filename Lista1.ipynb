### b)


O sinal é definido como:


x(t) =
\begin{cases}
t - 1, & 1 \le t < 2, \\[6pt]
0, & \text{caso contrário.}
\end{cases}


A energia do sinal é dada por:

$$
E = \int_{-\infty}^{\infty} |x(t)|^2 \, dt
$$

Como \( x(t) \neq 0 \) apenas no intervalo \( 1 \le t < 2 \):
$$
E = \int_{1}^{2} (t - 1)^2 \, dt
$$

Calculando a integral:

$$
E = \left[ \frac{(t - 1)^3}{3} \right]_{1}^{2}
$$

$$
E = \frac{(2 - 1)^3}{3} - \frac{(1 - 1)^3}{3}
$$

$$
\boxed{E = \tfrac{1}{3}}
$$



# Questão 2
### a)
temos o sinal x(t) como:
$$ 
x(t) = 
\begin{cases}
0, & t < -1.5, \\
2, & -1.5 \le t < 0, \\
2e^{-t/2}, & 0 \le t < 3, \\
0, & t \ge 3.
\end{cases}
$$ 
função degrau unitario:
$$
x(t) = 2 \big[u(t + 1.5) - u(t)\big] + 2 e^{-t/2} \big[u(t) - u(t - 3)\big]
$$




### b)

```python
import numpy as np
import matplotlib.pyplot as plt

def X_of_t(t):
    t = np.array(t)
    x = np.zeros_like(t, dtype=float)
    # 2 entre -1.5 e 0
    mask1 = (t >= -1.5) & (t < 0)
    x[mask1] = 2
    # exponencial entre 0 e 3
    mask2 = (t >= 0) & (t < 3)
    x[mask2] = 2.0 * np.exp(-t[mask2] / 2.0)
    return x

t = np.linspace(-6, 8, 2801)
x = X_of_t(t)
x_neg = X_of_t(-t)
x_t_plus2 = X_of_t(t+2)
x_t_minus3 = X_of_t(t-3)
x_2t = X_of_t(2*t)
x_t_div3 = X_of_t(t/3)

plt.figure(figsize=(14,10))

plt.subplot(3,2,1)
plt.plot(t, x, linewidth=1.5)
plt.title('x(t)')
plt.grid(True)
plt.axhline(0,color='k',linewidth=0.5)

plt.subplot(3,2,2)
plt.plot(t, x_neg, linewidth=1.5)
plt.title('x(-t)')
plt.grid(True)
plt.axhline(0,color='k',linewidth=0.5)

plt.subplot(3,2,3)
plt.plot(t, x_t_plus2, linewidth=1.5)
plt.title('x(t + 2)')
plt.grid(True)
plt.axhline(0,color='k',linewidth=0.5)

plt.subplot(3,2,4)
plt.plot(t, x_t_minus3, linewidth=1.5)
plt.title('x(t - 3)')
plt.grid(True)
plt.axhline(0,color='k',linewidth=0.5)

plt.subplot(3,2,5)
plt.plot(t, x_2t, linewidth=1.5)
plt.title('x(2t)')
plt.grid(True)
plt.axhline(0,color='k',linewidth=0.5)

plt.subplot(3,2,6)
plt.plot(t, x_t_div3, linewidth=1.5)
plt.title('x(t/3)')
plt.grid(True)
plt.axhline(0,color='k',linewidth=0.5)

plt.tight_layout()
plt.show()
```

# Questão 3

```python
import numpy as np
import matplotlib.pyplot as plt
# parametros
#frequencia da mensagem
fm = 50.0 
#freuqencia da portadora 
fc = 500.0 
#amplitude dc
A = 1.5
#frequencia de amostragem
fs = 20000
t = np.linspace(0, 0.1, int(0.1 * fs))

 #sinal da mensagem
x = np.cos(2 * np.pi * fm * t)
#sinal modulado
y = (A + x) * np.cos(2 * np.pi * fc * t)
 #plot 
plt.figure(figsize=(8,3))
plt.plot(t, y)
plt.title("Gráfico")
plt.xlabel("t (s)")
plt.ylabel("y(t)")
plt.xlim(0, 0.02)  # zoom de 20 ms para ver a portadora
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

u = lambda t: np.where(t >= 0, 1.0, 0.0)
t = np.linspace(0, 10, 2001)
dt = t[1] - t[0]

# Sinais
x = np.exp(-t) * u(t)
h = np.exp(-2 * t) * u(t)

# Convolução contínua aproximada
y = np.convolve(x, h)[:len(t)] * dt

# Plot
plt.figure(figsize=(8,3))
plt.plot(t, y)
plt.title("Gráfico")
plt.xlabel("t (s)")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```
